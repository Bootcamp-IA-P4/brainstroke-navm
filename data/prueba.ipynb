```python
# Cargar el dataset original
df1 = pd.read_csv('dataset_encoded.csv')

# Hacer una copia para trabajar
f1 = df1.copy()
f2 = df2.copy()
```

```python

f2_filtrado = f2.rename(columns={
    'Age': 'age',
    'BMI': 'bmi',
    'Sex': 'gender',
    'HighBP': 'hypertension',
    'HeartDiseaseorAttack': 'heart_disease',
    'Stroke': 'stroke'
})


columnas_comunes = ['age', 'bmi', 'gender', 'hypertension', 'heart_disease', 'stroke']
f1_filtrado = f1[columnas_comunes]
f2_filtrado = f2_filtrado[columnas_comunes]


df_concatenado = pd.concat([f1_filtrado, f2_filtrado], ignore_index=True)


print(df_concatenado['stroke'].value_counts())
```

```python
df_concatenado.to_csv('dataset_concatenado.csv', index=False)
```
